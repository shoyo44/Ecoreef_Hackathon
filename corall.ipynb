```python
df.columns
df['Protected_Area (Yes/No)']=df['Protected_Area (Yes/No)'].map({'Yes':1,'No':0})
df.head()
```

```python
def category(s):
    if s>=50:
        return "Healthy"
    elif 30<=s<50:
        return "Partially Bleached"
    else:
        return "Bleached"
df['Coral_Health']=df['Coral_Cover (%)'].apply(category)
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='Coral_Health',data=df,palette='viridis',hue='Coral_Health',legend=False)
plt.title("Coral Health Distribution")
plt.show()
```

```python
import folium
import webbrowser
from branca.element import Element
from jinja2 import Template
m=folium.Map(location=[df['Latitude'].mean(),df['Longitude'].mean()],zoom_start=8)

for _,row in df.iterrows():
    color='green' if row['Coral_Health']== 'Healthy' else 'yellow' if row['Coral_Health']=='Partially Bleached' else 'red'
    folium.CircleMarker(
        location=[row['Latitude'],row['Longitude']],
        radius=5,
        color=color,
        fill=True,
        fill_opacity=0.7  
    ).add_to(m)
legend_html = '''
<div style="
    position: fixed; 
    bottom: 50px; left: 50px; width: 180px; height: 140px; 
    background-color: white; z-index:9999; 
    font-size:14px;
    border:2px solid grey; 
    padding: 10px;">
    <b>Coral Health Legend</b> <br>
    <i class="fa fa-circle" style="color:green"></i> Healthy Coral <br>
    <i class="fa fa-circle" style="color:yellow"></i> Partially Bleached <br>
    <i class="fa fa-circle" style="color:red"></i> Bleached Coral
</div>
'''
m.get_root().html.add_child(Element(legend_html))
file="output.html"
m.save(file)
webbrowser.open(file)
    
    
```

```python
X=df.drop(['Reef_ID','Location','Coral_Health'],axis=1)
plt.figure(figsize=(10,8))
sns.heatmap(X.corr(),cmap='coolwarm',annot=True)
plt.title('Feature Correlation Map')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
features=['Water_Temp (°C)', 'Salinity (ppt)', 'pH', 'Turbidity (NTU)',
       'Nutrient_Levels (µmol/L)', 'Fish_Diversity (Species Count)',
       'Bleaching_Events (Count)', 'Human_Impact_Index',
       'Protected_Area (Yes/No)']
y=df['Coral_Health']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,accuracy_score
model=RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(X_train_scaled,y_train)
y_pred=model.predict(X_test_scaled)
print("Accuracy: ",accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
import pandas as pd
from flask import Flask, request, jsonify
import folium
try:
    df = pd.read_csv("coral_reef.csv")
except FileNotFoundError:
    print("Error: coral_reef.csv not found. Please ensure the file is in the correct path.")
    exit()

# Initialize Flask App
app = Flask(__name__)

# Function to generate coral reef map
def generate_map():
    m = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=5)
    for _, row in df.iterrows():
        color = 'green' if row['Coral_Cover (%)'] > 50 else 'yellow' if row['Coral_Cover (%)'] > 20 else 'red'
        folium.CircleMarker(
            location=[row['Latitude'], row['Longitude']],
            radius=5,
            color=color,
            fill=True,
            fill_opacity=0.7,
            popup=f"Location: {row['Location']}\nCoral Cover: {row['Coral_Cover (%)']}%"
        ).add_to(m)
    
    # Save map as HTML and return link
    map_file = "coral_map.html"
    m.save(map_file)
    return jsonify({"reply": "Map generated successfully. Open 'coral_map.html' to view the map."})

# Chatbot Route
@app.route('/chat', methods=['POST'])
def chatbot():
    user_input = request.json.get('message', '').lower()
    
    if "hello" in user_input or "hi" in user_input:
        return jsonify({"reply": "Hello! How can I assist you with coral reef insights today?"})

    if "coral health" in user_input:
        return jsonify({"reply": "Healthy corals are shown in green, partially bleached in yellow, and bleached in red."})
    
    elif "show map" in user_input:
        return generate_map()
    
    elif "protected areas" in user_input:
        protected = df[df['Protected_Area (Yes/No)'] == 'Yes']
        locations = protected[['Location', 'Latitude', 'Longitude']].to_dict(orient='records')
        return jsonify({"reply": f"Protected areas are present in: {', '.join(protected['Location'].unique())}", "locations": locations})

    else:
        return jsonify({"reply": "I'm here to help with coral reef data, maps, and environmental insights. Try asking about coral health or protected areas!"})

if __name__ == "__main__":
    app.run(debug=True, port=5001)  # Custom port for better control
```
